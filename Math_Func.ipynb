# Algebraic Functions
An algebraic function is a type of function that can be defined using polynomials. In other words, it is a function that involves only algebraic operations (addition, subtraction, multiplication, division) and can also involve the extraction of roots. It can be expressed using a finite number of these operations.

## Lineal function

It has the shape $$f(x)=mx + b$$ where $m$ y $b$  $\in R$. The gradient $m$  can be calculated by: $$m=\frac{y_{2}-y_{1}}{x_{2}-x_{1}}$$

and $b$ is the cut point with the axis $y$. The domain is $Dom_{f} = (-\infty, \infty)$. The range is $Im_{f} = (-\infty, \infty)$

```python
#Implementation of a linear function
N = 100 #Number of points of discrete time

m = 0 #Gradient

b = 0 # cut point with axis y

def f(x):
  m = int(input("Value of gradient m? "))
  b = int(input("Value of cut point b? "))
  print(f"The function is: f(x) = ({m})x+({b})")
  return m*x+b

x = np.linspace(-10,10, num=N)
y = f(x)
fig, ax = plt.subplots()
ax.plot(x,y)
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_title("Linear function")
ax.grid()
plt.show()
```

### Power functions

There are some functions that are a particular case of polynomial functions that are power functions, which have the form:


$$f(x)= x^{a}, a \in R$$


**For example:**

$$f(x)= x^{2}$$

The domain of $f(x)=x^{2}$ is $Dom_{f} = (-\infty, \infty)$. Its image is $Im_{f} = [0, \infty)$

```python
def f(x,power):
  return x**power

power = int(input("Exponent? "))
y = f(x,power)

fig,ax = plt.subplots()
ax.plot(x,y)
ax.grid()
ax.set_title("Exponential function")
ax.set_xlabel("x")
ax.set_ylabel("y")
plt.show()
```

### Logarithm function
The logarithm is defined by the **ratio**:

$$log_{b}(x) = n \Longleftrightarrow x=b^n$$

where:



* $b$ is the base.
* $n$ is the exponent to which the base is raised.
* $x$ is the result of raising the base $b$ to the exponent $n$

**Example:**

Taking b=2 and n=8, then:

$$2^8=256$$

So $x=256$. Calculating the base 2 logarithm of $x$ is:

$$log_{2}(256) = 8$$

```python
def f(x):
  return np.log2(x)

x = np.linspace(0.001,256, num=1000)

plt.grid()
plt.xlabel("x")
plt.ylabel('y')
plt.title("Logarithm function")
plt.plot(x,f(x))
plt.show()
```

# Sectioned function

They are functions that have different values defined by an interval. For example, the Heaviside step function:

$$H(x) =
      \begin{cases}
         0, &\quad \text{for, } x < 0 \\
         1, &\quad\text{for. } x \ge 0 \\
      \end{cases}
$$

```python
def H(x):
  Y = np.zeros(len(x))
  for idx,x in enumerate(x):
    if x>=0:
      Y[idx]=1
  return Y

N=1000
x = np.linspace(-10,10, num=N)
y = H(x)

plt.grid()
plt.xlabel('x')
plt.ylabel('y')
plt.title("Step function")
plt.plot(x,y)
plt.show()
```

Another sectioned function is the absolute value. This function only returns positive values.
$$F(x) =
      \begin{cases}
          \ x, &\quad \text{for. } x > 0 \\
        -x, &\quad\text{for. } x < 0 \\
         0, &\quad\text{for. } x=0
      \end{cases}
$$

```python
def abs(x):
  y = np.zeros(len(x))
  for i,x in enumerate(x):
    if x<0:
      y[i] = -1*x
    else:
      y[i] = x
  return y

if __name__=="__main__":
  y = abs(x)
  plt.grid()
  plt.xlabel('x')
  plt.ylabel('y')
  plt.title("Absolute function")
  plt.plot(x,y)
  plt.show()
```

#Composite functions
A composite function is generally a function that is written inside another function. Composition of a function is done by substituting one function into another function.
$$
(fÂ·g)(x) = f[g(x)]
$$

$(fÂ·g)(x)$ is not the same as $(gÂ·f)(x)$


```python
N = 1000
x = np.linspace(-10,10,num=N)
def g(x):
  return x**2

def f(x):
  return np.sin(x)

y = g(x)
f_o_g = f(g(x))

plt.grid()
plt.xlabel('x')
plt.ylabel('y')
plt.title("Composite function")
plt.plot(x,f_o_g)
plt.show()
```

#How to manipulate functions?

### Vertical and horizontal displacements

Being $c$ a constant greater than zero, then the graph:

- $y=f(x)+c$ shifts $c$ units up.
- $y=f(x)-c$ shifts $c$ units down.
- $y=f(x-c)$ shifts $c$ units to the right.
- $y=f(x+c)$ shifts $c$ units to the left.

```python
N = 1000

#Define a second grade function
def f(x):
  return x**2;

'''You can assign the c value and replace 'y' in other form
for example:
  y = f(x)+c
  y = f(x)-c
  y = f(x+c)
  y = f(x-c)
'''
c = 5

x = np.linspace(-10,10, num=N)

y = f(x+c)

fig, ax = plt.subplots()
ax.plot(x,y)
ax.grid()
ax.set_title("Manipulating a second degree function")
ax.axhline(y=0, color='r')
ax.axvline(x=0, color='r')
```

### Elongations and compressions

Being $c$ a constant greater than zero, then the graph:

- $y=c \cdot f(x)$ stretches the graph vertically by a factor of $c$.
- $y= \frac{1}{c} \cdot f(x)$ compresses the graph vertically by a factor of $c$.
- $y=f(c \cdot x)$ compresses the graph horizontally by a factor of $c$.
- $y= f(\frac{1}{c} \cdot x )$ stretches the graph horizontally by a factor of $c$.

```python
N = 1000

def f(x):
  return np.sin(x);

c = 2

x = np.linspace(-15,15, num=N)

y = f((1/c)*x)


fig, ax = plt.subplots()
ax.plot(x,y)
ax.grid()
ax.axhline(y=0, color='r')
ax.axvline(x=0, color='r')
```

### Reflections

- $y=-f(x)$ reflects the graph about the x-axis.
- $y=f(-x)$ reflects the graph about the y axis.

```python
N = 1000

def f(x):
  return x**3;

x = np.linspace(-10,10, num=N)

y = f(-x)

fig, ax = plt.subplots()
ax.plot(x,y)
ax.grid()
ax.axhline(y=0, color='r')
ax.axvline(x=0, color='r')
```

### Odd function
A function is odd if it satisfies the following relationship along its domain: \
$f(-x) = -f(x)$ \

This relationship tells us that a function is odd if it is symmetric to the horizontal axis (X axis). For example, a cubic function is odd.

The two previous characteristics or properties are closely related to symmetry. You can test them yourself with different functions and even make a small Python routine that checks if a function is even or odd.

### Bounded function
A function is bounded if its codomain (also known as range or image) is between two values, that is, it is bounded. This definition is defined as that there is a number m that for every value of the domain of the function is fulfilled that: \
$-m \leq f(x) \leq m$ \
For example, the sine or cosine function is bounded on the interval [-1, 1] within its co-domain.
